### Data Visualization

<b>Scatterplot with python

```python
# create a figure and axis
fig, ax = plt.subplots()
# scatter the sepal_length against the sepal_width
ax.scatter(df['SepalLengthCm'], df['SepalWidthCm'])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
# create color dictionary
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g',
'Iris-virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(df['SepalLengthCm'])):
    ax.scatter(df['SepalLengthCm'][i],
    df['SepalWidthCm'][i],color=colors[df['Species'][i]])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

<b>Line Chart

```python
# get columns to plot
columns = df.columns.drop(['Species'])
# create x data
x_data = range(0, df.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, df[column], label=column)
# set title and legend
ax.set_title('Iris Dataset')
ax.legend()
```

### Scatter Plot- Distribution & Relationship with Pandas

```python
plt.figure(figsize=(8,4))
plt.scatter(x=df['SepalLengthCm'], y= df["SepalWidthCm"])
plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(8,4))
df.plot(kind ='scatter', x='PetalLengthCm', y = 'PetalWidthCm')

plt.show()
```

### Displot-kDeplot-Boxplot

```python
sns.distplot(df['SepalLengthCm'])
plt.show()
```

```python
sns.kdeplot(df['SepalLengthCm'], shade = True, hue= df['Species'])
plt.show()
```

```python
sns.boxplot(data= df)
plt.tight_layout
plt.show
```

```python
sns.boxplot(data= df, x='Species', y='PetalWidthCm')
plt.xlabel('Species')
plt.ylabel('PetalWidth')
plt.show()
```
