# Drop kolom-kolom yang mengandung missing value lebih dari 50% keseluruhan data.

```python
data.isnull().mean()
```

```python
missing_value = data.columns[data.isnull().mean()>0.5]
missing_value
```

```python
data.drop(missing_value, axis=1, inplace= True)
data.isnull().mean()
```

# Isi missing value pada kolom country dan children dengan data yang paling banyak muncul.

```python
data.loc[: ,["country","children"]]
```

```python
data.country.value_counts()
```

```python
data.country.isnull().sum()
```

```python
Vcountry = data.country.value_counts().idxmax()
Vcountry

```

```python
data.country.fillna(Vcountry, inplace=True)
data.country.isnull().sum()
```

```python
data.info()
```

```python
data.country.value_counts() #Memastikan bahwa yang dimasukkan adalah data yang paling banyak muncul
```

```python
data.children.value_counts()
```

```python
Vchildren = data.children.value_counts().idxmax()
Vchildren
```

```python
data.children.fillna(Vchildren, inplace=True)
data.children.isnull().sum()
```

```python
data.children.value_counts() #Memastikan bahwa yang dimasukkan adalah data yang paling banyak muncul
```
