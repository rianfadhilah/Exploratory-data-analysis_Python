### Titanic Dataset Visualization.

Dataset yang digunakan adalah dataset Titanic (https://www.kaggle.com/competitions/titanic/data). Berdasarkan dataset ini, saya melakukan visualisasi data dengan dari beberapa macam variabel,

Single Variable
Two Variable
Multivariable

<b>Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
df = pd.read_csv('titanic.csv')
df.head(10)
```

```python
df.tail(10)
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df['Age'] =df['Age'].astype("int")
```

```python
df.columns
```

<b>Single Variable

```python
plt.figure(figsize = (8,4))
plt.hist(df['Age'])
plt.title("frequency distributions of age in titanic ship")
plt.xlabel("Age")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

Distribusi Umur pada dataset titanic memiliki kecenderungan ke kiri dimana itu menggambarkan mayoritas penumpang berada pada kisaran umur yang masih muda dengan mayoritas umur berada pada rentang umur 20 sampai 30 tahun.

```python
dfData = df[[ 'Fare']]
dfData.head()
```

```python
sns.boxplot(data= dfData)
plt.tight_layout
plt.show
```

Dari dataset dapat diketahui bahwa pada kolom fare atau tarif, terdapat banyak outlier  yang memiliki harga yang sangat tinggi dibandingkan median yang terdapat pada dataset, dengan extream value nya memiliki nilai tertinggi diatas 500

```python
sns.boxplot(x = "Sex", y = "Age", data = df)
```

Jenis kelamin laki-laki dan perempuan memiliki median umur yang hampir sama dengan jenis kelamin laki-laki memiliki beberapa penumpang berusia lebih dari 60 tahun.

```python
Valueclass= df.Pclass.value_counts()
Valueclass
```

```python
fig, ax = plt.subplots(figsize=(30,6))
explode = (0.1, 0, 0,)
Valueclass.plot(kind="pie",autopct= '%1.0f%%',explode=explode,shadow=True)
plt.tight_layout()
plt.legend()
ax.set_title('3 class of Titanic Ship')
plt.show
```

Pada dataset titanic terdapat 3 kategori kelas, dimana 55% penumpang berada pada kelas ke tiga yaitu kelas paling murah, dengan kelas 1 yang mana merupakan kelas termahal menjadi kelas terbanyak 2 dengan 24% dan kelas kedua menjadi yang terakhir dengan 21% dari total penumpang

<b> Two Variable

```python
dg = df.groupby(['Sex'])['Survived'].value_counts().reset_index(name='counts')
dg
```

```python
ax = sns.histplot(
    dg,
    x='Survived',
    weights='counts',
    hue='Sex',
    bins=10,
    multiple='stack',
    palette=['#24b1d1', '#ae24d1'],
    edgecolor='white',
    
)
ax.set_xticks(range(0,2))
plt.tight_layout
```

Berdasarkan data dari dataset dapat dilihat bahwa sebanyak lebih dari 500 orang tidak selamat saat tenggelamnya titanic dengan mayoritas penumpang yang tidak selamat adalah laki-laki, sedangkan lebih dari 300 penumpang yang selamat sebagian besar adalah perempuan.

```python
df1_plot = df.groupby(['Sex', 'Survived']).size().reset_index().pivot(columns='Sex', index='Survived', values=0)
df1_plot.plot(kind='bar', stacked=True)
```

```python
df_plot = df.groupby(['Pclass', 'Survived']).size().reset_index().pivot(columns='Pclass', index='Survived', values=0)
df_plot.plot(kind='bar', stacked=True)
```

Berikut adalah visualisasi dimana penumpang yang paling banyak tidak selamat adalah penumpang yang berada pada kelas ke 3 yaitu kelas yang paling murah, sedangkan untuk penumpang yang selamat sedikit didominasi oleh penumpang kelas 1 yaitu kelas yang paling mahal

```python
plt.figure(figsize = (8,4))
sns.countplot(data=df, x='Survived', hue='Sex')
plt.grid(True)
plt.title("Comparison survivor of titanic")
```

```python
sns.scatterplot(y = df['Parents/Children Aboard'], x = df['Fare'], hue = df['Pclass'], palette=sns.color_palette('bright')[:3])
```

Semakin banyak orang tua atau anak yang dibawa oleh penumpang, semakin mungkin mereka membeli tiket kelas 3.

<b> Multi Variable

```python
sns.pairplot(df, hue='Survived', palette=sns.color_palette("bright")[:2])
```

Pada visualisasi diatas dapat dilihat bahwa semakin banyak penumpang yang memiiki saudara ataupun pasangan maka penumpang tersebut akan semakin mungkin tidak selamat saat tenggelamnya kapal titanic, begitu juga dengan penumpang yang membawa orang tua atau anak, semakin banyak penumpang membawa orang tua atau anak semakin besar kemungkinan mereka tidak selamat.
